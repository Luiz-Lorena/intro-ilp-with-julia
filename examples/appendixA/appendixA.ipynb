## Advanced example

In this subsection we are going to use Julia to implement the Simulated Annealing (SA) metaheuristic to approximately solve the Traveling Salesman Problem.

```julia
using Random         # For reproducibility 
using DelimitedFiles # To read CSV files
using Distances      # For distance calculations
using Plots          # For plotting the route

# Structure to hold the problem data
struct ProblemData
  distance_matrix # Distance matrix
  coordinates     # Vector of coordinates 
end

# Structure to hold the solution
mutable struct Solution
  route     # Vector of city indices
  objective # Objective value
end

# Function to plot the route of a solution 
function plot_route(s, data)
  # Base plot: the cities
  p = plot(data.coordinates[:, 1], 
           data.coordinates[:, 2],
           seriestype = :scatter,
           label = "Cities",
           legend = :topleft)
  # Add label for each city
  for i in 1:length(sBest.route)
    x = data.coordinates[sBest.route[i], 1]
    y = data.coordinates[sBest.route[i], 2]
    annotate!(x, y + 0.0005, 
              text(string(sBest.route[i]), 8))
  end
  # Extract ordered route coordinates
  xs = data.coordinates[sBest.route, 1]
  ys = data.coordinates[sBest.route, 2]
  # Include the starting point (loop)
  xf = vcat(xs, xs[1])
  yf = vcat(ys, ys[1])
  # Draw lines between the cities
  plot!(xf, yf, seriestype=:path, 
        color=:red, label="Best Route", lw=2)
  # Show the plot
  display(p)
end

# Function to calculate the objective value
function f(s, data)
  s.objective = 0.0
  n = length(s.route)
  for i in 1:n-1
    s.objective += data.distance_matrix[s.route[i], s.route[i+1]]   
  end
  # Add distance from last to first city
  s.objective += data.distance_matrix[s.route[n], s.route[1]]
end

# Neighbor function (Swap two cities)
function N(s, data)
  # Create a new solution based on the current one
  sNew = deepcopy(s)              
  n = length(s.route)
  # Randomly select two indices
  i, j = rand(1:n), rand(1:n) 
  while i == j
    j = rand(1:n)
  end
  # Swap the cities
  sNew.route[i], sNew.route[j] = sNew.route[j], sNew.route[i] 
  # Recalculate the objective value for the new solution
  f(sNew, data)
  return sNew
end

# Simulated Annealing algorithm for TSP
function SA(s, T, Tmin, α, L, data)
  sBest = deepcopy(s)
  while T > Tmin
    for l in 1:L
      # Generate a neighbor solution by swapping two cities
      sNew = N(s, data)
      # Calculate the change in objective value
      Δ = sNew.objective - s.objective
      # If the new solution is better or accepted by probability
      if Δ < 0 || rand() < exp(-Δ / T)
        s = deepcopy(sNew)
        if s.objective < sBest.objective
          sBest = deepcopy(s)
        end
      end
    end
    T *= α  # Cool down the temperature
  end
  return sBest
end

# Load the problem data from CSV files
coordinates = readdlm("data/tsp_dataset.csv", ',')

# Create a distance matrix from the coordinates using Distances.jl
distance_matrix = pairwise(Haversine(6371.0), coordinates; dims = 1)

# Initialize the problem data structure
data = ProblemData(distance_matrix, coordinates)

# Create an initial solution with a random route
s = Solution(shuffle!(collect(1:size(coordinates, 1))), Inf)

# Calculate the objective value for the initial solution
f(s, data)

# Run the Simulated Annealing algorithm
Random.seed!(42) # For reproducibility
sBest = SA(s, 1000.0, 100.0, 0.95, 10, data)

# Print the best solution found
println("Best route: ", sBest.route)
println("Best objective value: $(sBest.objective) km")

# Plot the best route
plot_route(sBest, data)
```
