```julia
# -------------------------------------------------------------
#  Code for: "Introduction to Integer Programming and Applications with Julia"
#  Chapter: 2 – Knapsack problems
#  Section: 2.2.1 – The Binary Knapsack Problem (BKP)
#  Author(s): Luiz Henrique Nogueira Lorena
# -------------------------------------------------------------

using JuMP  # Modeling language
using HiGHS # Solver

# Problem data
v = [6, 8]     # item value
w = [4, 3]     # item weight
W = 6          # knapsack capacity
n = length(w)  # number of items

# Binary model
model = JuMP.Model(HiGHS.Optimizer)

# Silent mode (solver output is not printed)
JuMP.set_silent(model)

# Decision variables
@variable(model, x[1:n], Bin)

# Objective
@objective(model, Max, sum(v[i] * x[i] for i=1:n))

# Constraint
@constraint(model, sum(w[i] * x[i] for i=1:n) <= W)

# Print model
print("Model:\n$model")

# Solve model
JuMP.optimize!(model)

# Display solution
println("\nSolution:")
for i in 1:n
    println("x[$i] = ", JuMP.value(x[i]))
end
println("\nObjective = ", JuMP.objective_value(model))
```
