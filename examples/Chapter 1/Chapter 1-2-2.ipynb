```julia
using JuMP  # Modeling language
using HiGHS # Solver

# Create the model
model = JuMP.Model(HiGHS.Optimizer)

# Silent mode (solver output is not printed)
JuMP.set_silent(model)

# Decision variables
@variable(model, 0 <= x1 <= 1)
@variable(model, 0 <= x2 <= 1)

# Objective
@objective(model, Min, 7x1 + 10x2)

# Constraint
@constraint(model, 3x1 + 5x2 >= 7)

# Optimize
JuMP.optimize!(model)

# Print model
print("Model:\n$model")

# Extract solution
println("\nSolution:")
println("x1 = ", JuMP.value(x1))
println("x2 = ", JuMP.value(x2))
println("Objective value = ", JuMP.objective_value(model))
```
